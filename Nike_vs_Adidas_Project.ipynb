```python
# Reading Discount data for both products from csv file
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np

#Reading csv file for Discount data of both brands
df = pd.read_csv('C:\\Users\\bhawn\\OneDrive\\Documents\\Bootcamp\\Discount_Data.csv')

#Fetching individual values of all columns
discount = df['Discount range'].values
names = df['Measure names'].values
values = df['Measure values'].values

#Summary Statstics
df.describe()

#Generating a crosstab
df

nikeArray =[]
nikeValues = []
adArray = []
adValues = []

#Creating data arrays for both the brands
for idx, val in enumerate(names):
    if val == "Nike" :
        nikeArray.append(val)
        nikeValues.append(values[idx])
    else :    
        adArray.append(val)
        adValues.append(values[idx])

#Fetching unique values of discount range column from csv data
uniqueDiscount = []
for i in discount:
    if i not in uniqueDiscount:
        uniqueDiscount.append(i)
        
# Setting the figure size
fig = plt.figure(figsize=(8, 8))       
x1 = np.arange(len(nikeArray))
w = 0.3
#Plotting bars for both brands and labeling it
plt.bar(x1, nikeValues, width=w, label='Nike')
plt.bar(x1+w, adValues, width=w, label='Adidas')
plt.xticks(x1, uniqueDiscount)
plt.ylim([0,1100])
plt.tight_layout()

# Labeling the X-axis
plt.xlabel('Discount Range(%)')

# Labeling the Y-axis
plt.ylabel('Number of Products')

# Giving a title to the graph
plt.title('Discount Statistics for both brands')

# Showing a legend on the plot 
plt.legend()

#Saving the plot as an image
fig.savefig('discount_data.jpg', bbox_inches='tight', dpi=150)

#Displaying the plot
plt.show()
```
